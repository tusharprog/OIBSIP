# Importing Liberaries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing dataset

```python
Unemployment_1 = pd.read_csv('Unemployment in India.csv')
Unemployment_2 = pd.read_csv('Unemployment_Rate_upto_11_2020.csv')
```

```python
Unemployment_1.head()
```

```python
Unemployment_2.head()
```

```python
Unemployment_1.count()
```

```python
Unemployment_1.info()
```

```python
Unemployment_1
```

```python
Unemployment_1.isnull().sum()
```

```python
Unemployment_1 = Unemployment_1.dropna()
```

```python
Unemployment_1.reset_index()
```

```python
Unemployment_2.isnull().sum()
```

```python
x = Unemployment_1.drop(['Area'], axis = 1)
x
```

```python
y = Unemployment_2.drop(['Region.1', 'longitude', 'latitude'], axis = 1)
y
```

```python
frames = [x, y]

Unemployment = pd.concat(frames)
```

```python
Unemployment.reset_index()
```

```python
Unemployment[' Frequency'].value_counts()
```

```python
Unemployment[' Frequency']=Unemployment[' Frequency'].replace(' Monthly','Monthly')
Unemployment[' Frequency']=Unemployment[' Frequency'].replace(' M','Monthly')
```

```python
Unemployment[' Frequency'].value_counts()
```

```python
Unemployment=Unemployment.rename(columns={' Date':'Date',' Frequency':'Frequency',' Estimated Unemployment Rate (%)':'Estimated Unemployment Rate (%)',' Estimated Employed': 'Estimated Employed',' Estimated Labour Participation Rate (%)':'Estimated Labour Participation Rate (%)'})
```

```python
Unemployment.info()
```

```python
Unemployment.head()
```

## Changing to format of Date from object to DataTime


```python
Unemployment["Date"] = pd.to_datetime(Unemployment["Date"])
```
